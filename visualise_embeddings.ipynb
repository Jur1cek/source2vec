```python
from gensim.models import KeyedVectors
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import umap

init_notebook_mode(connected=True)
```

Model loading - only top 2000 words

```python
model = KeyedVectors.load_word2vec_format("../model.vec", binary=False, limit=100)
```

Preprocess word dictionary

```python
tokens=[]
word_labels=[]

for word, vocab_obj in model.vocab.items():
    word_labels.append(word)
    tokens.append(model[word])
```

Dimensionality reduction via umap (https://arxiv.org/abs/1802.03426) (much faster than TSNE with comparable results)  
We need to project 128 dimensional vector to 2D space (X, Y coordinates)

```python
Y = umap.UMAP().fit_transform(tokens)
```

Scatter plot visualisation using plotly (be carefull how many words you want to visualise this way), you can kill your browser :))

```python
x_coords = Y[:, 0]
y_coords = Y[:, 1]

trace1 = go.Scatter(
    x=x_coords,
    y=y_coords,

    mode='text',
    text=word_labels,
    textfont=dict(
        size=9,
    )
)

data=[trace1]
fig = go.Figure(data=data)
iplot(fig)
```
